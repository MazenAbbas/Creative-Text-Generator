```python
from transformers import pipeline

# Load the text generation pipeline using the 'distilgpt2' model
generator = pipeline('text-generation', model='distilgpt2')

# Define the starting text
starting_text = "Once upon a time, in a land far away,"

# Generate text based on the starting text
generated_text = generator(starting_text, max_length=100, num_return_sequences=1)

# Print the generated text
print(generated_text[0]['generated_text'])
```

```python
from transformers import pipeline

# Load the text generation pipeline using the 'distilgpt2' model
generator = pipeline('text-generation', model='distilgpt2')

# Define the starting text
starting_text = "In a world where artificial intelligence rules,"

# Generate text with custom parameters
# We'll use different decoding strategies here for better control
generated_text_custom = generator(
    starting_text,
    max_length=150,
    num_return_sequences=3,  # Generate 3 different outputs
    do_sample=True,          # Use sampling for more creative output
    temperature=0.7,         # Controls the randomness (lower = more predictable)
    top_k=50,                # Filters out low-probability words
    top_p=0.95               # Another filtering method for diverse output
)

# Print the generated texts
for i, text_output in enumerate(generated_text_custom):
    print(f"--- Generated Text {i+1} ---")
    print(text_output['generated_text'])
    print("\n")
```
